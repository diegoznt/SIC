```python
#Números primos

def es_primo(n):
  if n < 2:
    return False
  for i in range(2,n):
     if n % i == 0:
                         # return False TAMBIEN PUEDE SER CON RETURN PARA TENER UN VALOR LOGICO
        return print("Su número no es primo:", False)
  return print("Su número es primo:", True)  #return True
  #return True


```

```python
es_primo(7)

```

# Sucesión de fibonacci

```python
def fibonacci(n):
  fib_seq = [0,1]
  for i in range(2,n):
    fib_seq.append(fib_seq[-1] + fib_seq[-2])
    print(fib_seq)
  return fib_seq

fibonacci(9)
```

```python
#Contar palabras de un texto

def contar(texto):
  palabras = texto.split() # Separa las palabras
  conteo = {}
  for i in palabras:
    conteo[i] = conteo.get(i,0) + 1 # nos da la palabra mas un 1 y lo guarda en conteo, nos da el conteo de palabras por aparición

  return conteo
```

```python
texto= "hola buenas tardes hola"
print(contar(texto))
a=contar(texto)
a['hola']

#texto.split()
```

```python
#Factorial
def factorial(n):
 if n==0 or n==1:
  return 1
 else:
  return n*factorial(n-1) # interesante como se multiplica por la misma funcion hasta 1 o cero la iteracion sin necesidad de for

```

```python
factorial(4)
```

# SINTAXIS INTERESANTE DE CICLO

```python
#Suma de los digitos

def suma_digitos(n):
  return sum(int(digito) for digito in str(n)) # convierte cada el numero entero a string y luego recorre cada digito y lo pasa a entero y los suma todos

suma_digitos(1234)
```

```python
# Mayor de tres números
def mayor_de_tres(a,b,c):
  return max(a,b,c)

mayor_de_tres(93280,6,1000)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Vamos a hacer una limpieza sacar NANs
flights.dropna(subset=['arr_delay','carrier','dest','tailnum'], inplace = True) # dropna(...) elimina filas que tienen valores vacíos (NaN).

#subset=[...] indica en qué columnas revisar si hay datos faltantes.

#inplace=True significa que los cambios se aplican directamente sobre el DataFrame flights (no se crea una copia).

```

```python
flights_airlines=flights.merge(airlines, on='carrier') # ahora combinamos ambos set pero con la base limpia
flights_airports = flights.merge(airports, left_on='dest',right_on='faa')
flight_planes = flights.merge(planes, on='tailnum', how='left') #flights	Es la tabla base con los datos de los vuelos.
#.merge(planes, ...)	Une los datos de vuelos con los datos de los aviones (de planes).
#on='tailnum'	La columna 'tailnum' está en ambos DataFrames y se usa como clave para unirlos.
#how='left'	Especifica un left join, es decir:
#→ Se conservan todas las filas de flights aunque no haya coincidencia en planes.
#→ Si no hay coincidencia, los valores de planes serán NaN.

flights_weather= flights.merge(weather, on=['year','month','day','hour','origin'], how='left') # Finalmente al decir left se conservan las lineas de flights y se van uniendo las de weather segun esta fija
print(flights)
print(weather)
print(flights_weather)
```

```python
# Las 5 aerolineas mas puntuales
puntuales=flights_airlines.groupby('name')['arr_delay'].mean().sort_values(ascending=True)
puntual=puntuales.head(10)
puntual
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=puntual.values,y=puntual.index,palette='Greens_d')
plt.title('Top 10 aerolineas con menor retraso promedio')
plt.xlabel('Retraso promedio (minutos)')
plt.ylabel('Aerolinea')
plt.show
```

```python
plt.figure(figsize=(10,6))
sns.barplot(puntual.T,palette='Greens_d') # si coloco solo puntual igual me lo da pero se ve mas desordenado
plt.title('Top 10 aerolineas con menor retraso promedio')
plt.ylabel('Retraso promedio (minutos)')
plt.xlabel('Aerolinea')
plt.xticks(rotation=45)
plt.show
```

```python
# Ahora vamos a ver los con menos retrasos

flights_airports['retrasos'] = flights_airports['arr_delay']>15
flights_airports
retrasos=flights_airports.groupby('name')['retrasos'].sum().sort_values(ascending=False).head(5)
retrasos # nos da la cantidad de veces que se retrasaron mayor a 15 minutos
```

```python

plt.figure(figsize=(10,6))
sns.barplot(x=retrasos.values,y=retrasos.index,palette='Reds_d')
plt.title('Top 10 aerolineas con mas retrasos mayores a 15 minutos')
plt.xlabel('Cantidad de retrasos (minutos)')
plt.ylabel('Aerolinea')
plt.show
```

```python
# Vamos a ver la relacion entre vuelos retrasados y la temperatura
# Interesante que solo necesitamos el nombre del set de datos y el titulo de la columna y podemos plotear ambos
plt.figure(figsize=(10,6))
sns.scatterplot(data=flights_weather, x='temp',y='arr_delay',alpha=0.2)
plt.title('Retrasos según temperatura')
plt.ylabel('Tiempo de retraso (minutos)')
plt.xlabel('Temperatura (F°)')
plt.show

# Ahora con velocidad del viento
plt.figure(figsize=(10,6))
sns.scatterplot(data=flights_weather, x='wind_speed',y='arr_delay',alpha=0.2)
plt.title('Retrasos según Velocidad del viento')
plt.ylabel('Tiempo de retraso (minutos)')
plt.xlabel('Velocidad del viento (mph)')
plt.show

```
